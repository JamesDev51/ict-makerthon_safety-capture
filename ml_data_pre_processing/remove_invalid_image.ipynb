```python
import os

#  
txt_file_path="valid_filename.txt"
directory="D:/aihub_data/construction_site_secure_equipment_data/images/Training/busan_opera"

# txt      .
with open(txt_file_path, 'r') as f:
    valid_filename_list = f.read().splitlines()
valid_filename_set=set(valid_filename_list)

for filename in os.listdir(directory):
    image_path = os.path.join(directory, filename)
    #     ,  .
    if os.path.isfile(image_path) and filename not in valid_filename_set:
        os.remove(image_path)
        print(f"Deleted {image_path}")
    else:
        print(f"{image_path} is valid")
```
